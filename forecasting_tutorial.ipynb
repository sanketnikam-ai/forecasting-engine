# Forecasting Engine - Tutorial
This notebook demonstrates how to use the Forecasting Engine with all 6 methods.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from forecasting_engine import ForecastingEngine

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (15, 6)
```

## 1. Load Your Data

```python
# Load from CSV
df = pd.read_csv('your_data.csv')

# Or use sample data
# from forecasting_engine import create_sample_data
# df = create_sample_data()

print(f"Data shape: {df.shape}")
df.head()
```

## 2. Initialize the Forecasting Engine

```python
# Initialize engine
engine = ForecastingEngine(
    df=df,
    date_col='date',
    value_col='value',
    freq='D'  # D=daily, W=weekly, M=monthly
)

# Split into train/test (80/20 split)
engine.split_data(test_size=0.2)

print(f"Train size: {len(engine.train)}")
print(f"Test size: {len(engine.test)}")
```

## 3. Visualize the Data

```python
plt.figure(figsize=(15, 6))
plt.plot(engine.train.index, engine.train[engine.value_col], label='Train', linewidth=2)
plt.plot(engine.test.index, engine.test[engine.value_col], label='Test', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Data - Train/Test Split')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## 4. Run All Forecasting Methods

```python
# Run all available methods
results = engine.run_all_forecasts()

print("\nForecasting complete!")
print(f"Methods executed: {len(results)}")
```

## 5. Compare Performance Metrics

```python
# Get metrics summary
metrics_df = engine.get_metrics_summary()
print("\nPerformance Metrics (sorted by MAPE):")
print(metrics_df)

# Visualize metrics
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

metrics_df['MAE'].plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title('Mean Absolute Error (MAE)', fontweight='bold')
axes[0].set_ylabel('MAE')
axes[0].tick_params(axis='x', rotation=45)

metrics_df['RMSE'].plot(kind='bar', ax=axes[1], color='lightcoral')
axes[1].set_title('Root Mean Squared Error (RMSE)', fontweight='bold')
axes[1].set_ylabel('RMSE')
axes[1].tick_params(axis='x', rotation=45)

metrics_df['MAPE'].plot(kind='bar', ax=axes[2], color='lightgreen')
axes[2].set_title('Mean Absolute Percentage Error (MAPE)', fontweight='bold')
axes[2].set_ylabel('MAPE (%)')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## 6. Visualize All Forecasts

```python
plt.figure(figsize=(16, 8))

# Plot training data
plt.plot(engine.train.index, engine.train[engine.value_col], 
         label='Train', color='black', linewidth=2, alpha=0.7)

# Plot actual test data
plt.plot(engine.test.index, engine.test[engine.value_col], 
         label='Actual', color='blue', linewidth=2.5)

# Plot forecasts
colors = ['red', 'green', 'orange', 'purple', 'brown', 'pink']
for i, (method, result) in enumerate(engine.results.items()):
    forecast = result['forecast']
    plt.plot(engine.test.index[:len(forecast)], forecast,
             label=method, color=colors[i % len(colors)], 
             linestyle='--', linewidth=2, alpha=0.8)

plt.xlabel('Date', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.title('Forecasting Methods Comparison', fontsize=14, fontweight='bold')
plt.legend(loc='best', fontsize=11, framealpha=0.9)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 7. Get Best Performing Model

```python
best_model, best_metrics = engine.get_best_model(metric='MAPE')

print(f"🏆 Best Model: {best_model}")
print(f"\nMetrics:")
print(f"  MAE:  {best_metrics['MAE']:.3f}")
print(f"  RMSE: {best_metrics['RMSE']:.3f}")
print(f"  MAPE: {best_metrics['MAPE']:.2f}%")
```

## 8. Detailed View of Best Model

```python
# Plot only the best model vs actual
plt.figure(figsize=(15, 6))

plt.plot(engine.train.index, engine.train[engine.value_col], 
         label='Train', color='gray', linewidth=1.5, alpha=0.6)
plt.plot(engine.test.index, engine.test[engine.value_col], 
         label='Actual', color='blue', linewidth=2.5, marker='o')
plt.plot(engine.test.index[:len(engine.results[best_model]['forecast'])], 
         engine.results[best_model]['forecast'],
         label=f'{best_model} Forecast', color='red', linewidth=2.5, 
         linestyle='--', marker='x')

plt.xlabel('Date', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.title(f'Best Model: {best_model} (MAPE: {best_metrics["MAPE"]:.2f}%)', 
          fontsize=14, fontweight='bold')
plt.legend(loc='best', fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 9. Export Results

```python
# Export all forecasts to CSV
forecast_df = engine.export_results('forecast_results.csv')

print("Results exported to 'forecast_results.csv'")
print("\nPreview of results:")
forecast_df.head(10)
```
